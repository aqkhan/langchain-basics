## LLM setup

```python
from langchain.chat_models import ChatOpenAI
# Get OpenAI API key from the .env file
import os
from dotenv import load_dotenv
load_dotenv()

key = os.getenv("OPENAI_API_KEY")
llm = ChatOpenAI(
    temperature=0.1,
    model_name="gpt-3.5-turbo",
    openai_api_key=key
    )
```

### Basic prompt

```python
# Set the prompt
hard_prompt = "What are the 3 best destinations for people who like to party?"
# Prompt
print(llm.predict(hard_prompt))
```

### Generate embeddings

```python
from langchain.embeddings import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-ada-002", openai_api_key=key)

test_document = "I am some stupid test document residing in a PDF file or some other shit!"

query_result = embeddings.embed_query(test_document)

print(query_result)
```
